```python
list_items = soup.select("li.product.type-product")

known_currencies = {"$": "USD", "€": "EURO", "£": "GBP", "¥": "CNY"}  # Add more as needed
pattern = r"([$€£¥])([\d,.]+)\s*–\s*\1([\d,.]+)"  

data = []
for _ in list_items:
    rating_score = float(_.select("div.star-rating")[0].get("aria-label").split()[1])
    title = _.select("h2.woocommerce-loop-product__title")[0].text
    price = _.select("span.price")[0].text
    match = re.search(pattern, price)

    if match:
        currency_unit = known_currencies[match.group(1)]
        lowest_price = float(match.group(2).replace(",", ""))
        highest_price = float(match.group(3).replace(",", ""))
        
    data.append({
        "Title": title,
        "Rating score": rating_score,
        "Lowest price": lowest_price if lowest_price else None,
        "Highest price": highest_price if highest_price else None,
        "Currency unit": currency_unit if currency_unit else None
    })
        
    
```

```python
df = pd.DataFrame.from_dict(data=data)
product_names = df['Title'].to_list()
product_names
```

```python
from datetime import datetime

data = {}

for _ in product_names:
    
    def remove_special_character(file_path: str):
        file_path = file_path.lower()
        for _ in "()#%":
            file_path = file_path.replace(_, '')
        file_path = file_path.replace(' ', '-')
        return file_path
    
    html_file_path = os.path.join(folder_path, 'product', remove_special_character(_), 'home.html')
    soup = BeautifulSoup(open(html_file_path, encoding="utf8"), "html.parser")
    
    category = soup.find('nav', class_='woocommerce-breadcrumb').select('a')[1].text
    description = soup.select('div#tab-description')[0]
    reviews = soup.select('div#comments')[0]
    num_of_reviews = int(reviews.find('h2').text.split()[0])
    comments = [
        {
            "author": cmt.find('strong', class_='woocommerce-review__author').text,
            "published": datetime.strptime(cmt.find('time')['datetime'], "%Y-%m-%dT%H:%M:%S%z"),
            "rate": int(cmt.find('strong', class_='rating').text) if cmt.find('strong', class_='rating') else None
        }
        for cmt in reviews.find_all('li')
    ]
    data[_] = {
        "category": category,
        "description": description,
        "num_of_reviews": num_of_reviews,
        "comments": comments
    }
```

```python
df[['Category', 'Num of reviews']] = np.zeros
df['Category'] = df.apply(lambda row: data[row['Title']]['category'], axis=1)
df['Num of reviews'] = df.apply(lambda row: data[row['Title']]['num_of_reviews'], axis=1)
df
```
